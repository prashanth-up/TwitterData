```python

import twint
import csv
import pandas as pd
import time
```

```python

# setup twint
c = twint.Config()
c.Search = "Biden pardons marijuana"
c.Store_csv = True
c.Output = "biden_pardons_marijuana.csv"
c.Since = "2022-10-1"
c.Until = "2022-10-11"
c.Hide_output = True

# run search
twint.run.Search(c)
```

```python
# read csv file
df = pd.read_csv("biden_pardons_marijuana.csv")
```

```python
# make a dataframe 
df = pd.DataFrame(df)
```
